```python
plt.figure(figsize=(20, 8))
sns.set_style('darkgrid',
              {'axes.facecolor': '0.95',
               'grid.color': '0.1',
               'figure.facecolor': '0.95'})
plt.title('Who scored more: right-handed or left-handed?', fontsize=18, fontweight='bold')
plt.xlabel('Body parts', fontsize=18, fontweight='bold')
plt.ylabel('Goals', fontsize=18, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
sns.barplot(x=['Right foot', 'Left foot', 'Headers', 'Others'], y=[goals['right_foot'].sum(),
                                                                   goals['left_foot'].sum(),
                                                                   goals['headers'].sum(),
                                                                   goals['others'].sum()])
plt.show()
```

```python
plt.figure(figsize=(20, 8))
plt.title('Number of fouls by teams', fontsize=18, fontweight='bold')
sns.set_style('darkgrid',
              {'axes.facecolor': '0.95',
               'grid.color': '0.1',
               'figure.facecolor': '0.95'})
sns.barplot(x=disciplinary.groupby('club').sum()['fouls_committed'].index,
            y=disciplinary.groupby('club').sum()['fouls_committed'].values)

plt.xlabel('Club', fontsize=18, fontweight='bold')
plt.ylabel('Number of fouls', fontsize=18, fontweight='bold')
plt.xticks(fontsize=16, rotation=90)
plt.yticks(fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(20, 8))
plt.title('Who has the most number of completed passes?', fontsize=18, fontweight='bold')
sns.set_style('darkgrid',
              {'axes.facecolor': '0.95',
               'grid.color': '0.1',
               'figure.facecolor': '0.95'})
sns.barplot(x=distributon.sort_values(by='pass_completed', ascending=False)[0:61]['player_name'],
            y=distributon.sort_values(by='pass_completed', ascending=False)[0:61]['pass_completed'])
plt.xlabel('Football player', fontsize=18, fontweight='bold')
plt.ylabel('Number of completed pass', fontsize=18, fontweight='bold')
plt.xticks(fontsize=16, rotation=90)
plt.yticks(fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(20, 8))
plt.title('Average value of passing accuracy by teams', fontsize=18, fontweight='bold')
sns.set_style('darkgrid',
              {'axes.facecolor': '0.95',
               'grid.color': '0.1',
               'figure.facecolor': '0.95'})
sns.barplot(x=distributon.groupby('club').mean()['pass_accuracy'].sort_values(ascending=False).index,
            y=distributon.groupby('club').mean()['pass_accuracy'].sort_values(ascending=False).values)
plt.xlabel('Сlub', fontsize=18, fontweight='bold')
plt.ylabel('Average value of passing accuracy', fontsize=18, fontweight='bold')
plt.xticks(fontsize=16, rotation=90)
plt.yticks(fontsize=16)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 12))
ax.set_title('Defense play', fontsize=18, fontweight='bold')

width = 0.33
x = np.arange(31)
ax.bar(x - width/2.33, defending.sort_values('tackles', ascending=False)['t_won'][0:31], width=0.3, label='Won')
ax.bar(x + width/2.33, defending.sort_values('tackles', ascending=False)['t_lost'][0:31], width=0.3, label='Loss')
ax.set_xlabel('Football player', fontsize=18, fontweight='bold')
ax.set_ylabel('Tackles', fontsize=18, fontweight='bold')
ax.set_xticklabels(defending.sort_values('tackles', ascending=False)['player_name'][0:31], 
                   rotation=90, fontsize=18)
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
sns.set_style('darkgrid')
plt.legend(fontsize=18)
ax.set_xticklabels(defending.sort_values('tackles', ascending=False)['player_name'][0:31], 
                   rotation=90, fontsize=18)
ax.set_xticks(x, fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```
